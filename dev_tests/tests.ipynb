## xml.etree

https://docs.python.org/3/library/xml.etree.elementtree.html

```python
import xml.etree.ElementTree as ET
```

```python
GPX_FILE = "test_files/files/29_04_2023.gpx"
```

```python
name_space = {"topo": "http://www.topografix.com/GPX/1/1"}
```

```python
gpx_tree = ET.parse(GPX_FILE)
gpx_root = gpx_tree.getroot()
```

```python
gpx_root.tag
```

```python
gpx_root.attrib
```

```python
schema = gpx_root.get("{http://www.w3.org/2001/XMLSchema-instance}schemaLocation").split(" ")
schema.remove("")
schema
```

```python
for child in gpx_root:
    print(child.tag, child.attrib)
```

```python
metadata = gpx_root.findall("topo:metadata", name_space)

for data in metadata:
    link = data.find("topo:link", name_space)
    text = link.find("topo:text", name_space)
    time = data.find("topo:time", name_space)
    print(text.text, link.get("href"), time.text)
```

```python
tracks = gpx_root.findall("topo:trk", name_space)

for track in tracks:
    name = track.find("topo:name", name_space)
    segments = track.findall("topo:trkseg", name_space)
    print(name.text)
    for segment in segments:
        points = segment.findall("topo:trkpt", name_space)
        for point in points:
            elevation = point.find("topo:ele", name_space)
            time = point.find("topo:time", name_space)
            # print("TEST:", point.attrib)
            print(point.get("lat"), point.get("lon"), elevation.text, time.text)
```
